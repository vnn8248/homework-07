```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

response = requests.get("https://www.lemonde.fr/en/")
doc = BeautifulSoup(response.text)
```

```python
rows = []
items = doc.find_all("div", attrs={"class": "article"})

for item in items:
    row = {}
    title = item.select_one(".article__title")
    articleDesc = item.select_one(".article__desc")
    url = item.select_one("a")['href']
    img = item.select_one("img")
    premium = item.select_one(".icon__premium")
    byline = item.select_one(".article__byline")
    articleType = item.select_one(".article__type")
    
    print("------------------------------------------------------------------------------------------------------------")
    try:
        row["Title"] = title.get_text()
    except:
        print("Title not found")
        continue
    try:
        row["Description"] = articleDesc.get_text()
    except:
        print("Description not found")
    try:
        row["Article URL"] = url
    except:
        print("URL not found")
    try:
        row["Image URL"] = img["src"]
    except:
        print("No image found")
    if premium:
        row["Premium"] = "Premium"
    else:
        print("Not premium")
    try:
        row["Byline"] = byline.get_text()
    except:
        print("No byline found")
    try:
        row["Article Type"] = articleType.get_text()
    except:
        print("No Article Type")
    
    print(row)
    rows.append(row)

rows
```

```python
df = pd.DataFrame(rows)
df
```

```python
df.to_csv("leMonde-homepage-articles.csv", index=False)
```
